```python
from experta import *
```

```python
#Facts 16

class LeafFall(Fact):   #تساقط الأوراق
    pass
class DryBranch(Fact):  #تيبس الغصن
    pass
class Peacock(Fact): #ظهور بقع تشبه عين الطاووس على الأوراق
    pass 
class Coton(Fact):  # افرازات قطنية على الأوراق و على الثمار
    pass
class Tumor(Fact):   # وجود أورام على الأغصان
    pass
class RotFruits(Fact):  #تعفن الثمار 
    pass
class Green(Fact): #الثمار خضراء غير ناضجة 
    pass
class LeafSpot(Fact):  # وجود بقع على بنية على الوجه العلوي و بقع رمادية على الوجه السفلي
    pass
class Honey(Fact):     #افرازات عسلية
    pass
class RotRoot(Fact):   # تعفن الجذر
    pass 
class Larva(Fact):  # يرقة داخل اللب
    pass 
class Sawdust(Fact):  #نشارة خشب 
    pass
class PlantSap(Fact):  # عصارة نباتية
    pass
class DryFlower(Fact):  #جفاف الزهر
    pass
class YelloLineInLeaf(Fact): #خطوط صفراء داخل الأوراق
    pass
class DryOlive(Fact): #ثمرة الزيتون جافة
    pass
```

```python
class Olive(KnowledgeEngine):
    @Rule(NOT(LeafFall(W())))
    def askPapers(self):
        self.declare(LeafFall(input('Is Leaf fall (y/n) : ')))
    @Rule(LeafFall(W()))
    def askDryBranch(self):
        self.declare(DryBranch(input('Is Branh Dry (y/n) : ')))
        
        
    @Rule(LeafFall('y') & DryBranch('n'))
    def askPeacock(self):
        self.declare(Peacock(input('Is there a peacock on Leaf (y/n) : ')))
#   PEACOCK'S EYE OLIVE LEAF SPOT عين الطاووس
    @Rule(Peacock('y'))  
    def olive_peacock_spot(self):
        print('PEACOCK\'S EYE OLIVE LEAF SPOT ')
    @Rule(Peacock('n'))
    def askCototon(self):
        self.declare(Coton(input('Is there a Coton on paper (y/n) : ')))
#   OLIVE PSYLLID بسيلا الزيتون
    @Rule(Coton('y')) 
    def olivePsyllid(self):
        print('OLIVE PSYLLID ')
        
    @Rule(Coton('n'))
    def askForLeafSpot(self):
        self.declare(LeafSpot(input('Is There a Broun Spot on top of leaf and silver spot on down (y/n) : ')))
#   OLIVE LEAF SPOT تبقع زيتون
    @Rule(LeafSpot('y'))
    def leafSpot(self):
        print('OLIVE LEAF SPOT')
        
     
#   OLIVE SCALE INSECTS حشرات قشرية
    @Rule(LeafSpot('n')) 
    def leafSpot(self):
        print('OLIVE SCALE INSECTS')
    
    
    @Rule(LeafFall('y') & DryBranch('y'))
    def askHoney(self):
        self.declare(Honey(input('Is there honey substance on Leaf and olive (y/n) : ')))
#   BLACK SCALE نمشة سوداء
    @Rule(Honey('y'))   #
    def blackScale(self):
        print('BLACK SCALE')
       
    @Rule (Honey('n'))
    def askSawdust(self):
        self.declare(Sawdust(input('Sawdust under tree (y/n) : ')))
    @Rule(Sawdust('y'))
    def askPlantSap(self):
        self.declare(PlantSap(input('Abundant plant sap at the entrance holes ? (y/n) : ')))
#   OLIVE TREE BORER حفار الساق
    @Rule(PlantSap('y'))  
    def TREE_BORER(self):
        print('OLIVE TREE BORER')        
    
#   OLIVE BEETLE نيرون الزيتون
    @Rule(PlantSap('n'))  
    def PhloeotribusScarabaeoides(self):
        print('OLIVE BEETLE')
       
    @Rule(Sawdust('n')) 
    def asklineInLeaf(self):
        self.declare(YelloLineInLeaf(input('Is there a Yello Line In Leaf (y/n) : ')))
#   Acariens أكروسات
    @Rule (YelloLineInLeaf('n'))
    def acariens(self):
        print('Acariens')
       
    
            
    @Rule(LeafFall('n') & DryBranch('y'))
    def askTumor(self):
        self.declare(Tumor(input('Is there a tumor on Branch (y/n) : ')))
#   OLIVE KNOT  سل الزيتون
    @Rule(Tumor('y'))  
    def oliveKnot(self):
        print('OLIVE KNOT ')
       
#   WILT OLIVE TREE ذبول الزيتون
    @Rule(Tumor('n'))  
    def Wilt_olive_tree(self):
        print('WILT OLIVE TREE')
      

    @Rule(LeafFall('n') & DryBranch('n'))
    def askRotFruits(self):
        self.declare(RotFruits(input('Is Fruits Rot (y/n) : ')))
        
    
    @Rule(RotFruits('y'))
    def askDryOlive(self):
        self.declare(DryOlive(input('Is Olive Dry (y/n) : ')))
    @Rule(DryOlive('n'))
    def askLarva(self):
        self.declare(Larva(input('is there a Larvain inside olive')))
#   OLIVE FLY    ذبابة الزيتون
    @Rule(Larva('y'))   
    def oliveFly(self):
        print('OLIVE FLY')
        
       
    @Rule(Larva('n'))
    def askGreenFruits(self):
        self.declare(Green(input('Is Fruits Green or Mature (g/m) : ')))
#   OLIVE LEPROSYE تجعد الزيتون
    @Rule(Green('g')) 
    def wrinkle(self):
        print('OLIVE LEPROSYE')
   

#   Olive Anthracnose تقرح الزيتون
    @Rule(Green('m')) 
    def anthracnose(self):
        print('Olive Anthracnose')
        

    @Rule(RotFruits('n'))
    def askRotRoot(self):
        self.declare(RotRoot(input('Is Root Rot (y/n): ')))
#     Root Rot تعفن الجذر
    @Rule(RotRoot('y'))   
    def rotRoot(self):
        print('Root Rot')
        
        
    @Rule(RotRoot('n'))   
    def askDryFlower(self):
        self.declare(DryFlower(input('is Flower Dry (y/n) : ')))
    
#     OLIVE MOTH يرقة الزيتون
    @Rule(DryFlower('y') | YelloLineInLeaf('y') | DryOlive('y') )
    def oliveMoth(self): 
        print('OLIVE MOTH ')
    @Rule (DryFlower('n'))
    def connect(self):
        print('Connect with Expert To Laboratory analysis')

        
```

```python
olive = Olive()
olive.reset()
olive.run()
```
